```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Either through the "Object oriented approach" or the "pyplot/MATLAB approach". The "Object oriented approach" is the recommended approach, allowing more advanced plotting.
The "pyplot/MATLAB approach" can be used for quick plotting.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure is the whole figure and can be seen as a canvas.
Axes is the part of the canvas where the plot is attached.
Axis sets the scale and limits and generate ticks and ticklabels.
Artist is basically everything visible on the figure, including figure, axes and axis.

### When plotting in Matplotlib, what is the expected input data type?

numpy.array (or objects convertible to arrays)
numpy.asarray

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("y=x^2 Object-oriented approach")
ax.set_xlabel("x value")
ax.set_ylabel("y value")
```

```python
plt.plot(x, y)
plt.title("y=x^2 Pyplot approach")
plt.xlabel("x value")
plt.ylabel("y value")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))
axs[0].scatter(random_data_x, random_data_y)
axs[0].set_title("Scatter plot")
axs[1].bar(names, values)
axs[1].set_title("Bar plot")
```
